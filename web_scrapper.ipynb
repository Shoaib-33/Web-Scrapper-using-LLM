```python
def preprocess_text(text):
    # Remove unnecessary characters
    text = text.replace('\n', '').replace('\\', '')

    # Replace special characters
    text = text.replace('\'', "'").replace('"', '"')

    # Replace remaining escape characters
    text = text.replace('\\', '')

    return text

# Assuming c contains the text you provided

# Preprocess the text
preprocessed_text = preprocess_text(c)

# Optionally, split the text into smaller chunks or paragraphs if needed
# For example, splitting into paragraphs by detecting double line breaks
paragraphs = preprocessed_text.split('\n\n')

# Print or use the preprocessed text as needed
for paragraph in paragraphs:
    print(paragraph)
```

```python
def generate_content(c, schemas):
    results = {}
    for schema in schemas:
        prompt = ChatPromptTemplate.from_messages(
            [("human", f"You are an extractor.You will be given a string {c}.Your task is to extract the {schema} from the {c}.Dont output anything twice and every {schema} provide only the output what is mention in {schema}.The output in one {schema} shouldnot be repeated to another {schema}.Provide the output in JSON format.And dont write anything extra out of {schema}.")]
        )
        chain = prompt | chat
        content =""
        for chunk in chain.stream({"c": c}):  # Pass the HTML content as context
            content += chunk.content
        results[schema] = content.strip()
    return results

# Assuming you have 'c' containing the HTML page information
schemas = input("Enter the schemas you want to apply, separated by commas: ").split(',')
results = generate_content(c, schemas)
for schema, result in results.items():
    print(f"Schema: {schema}\nResult: {result}")
```
